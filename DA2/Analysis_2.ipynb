```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data2=pd.read_csv(r"C:\Users\LENOVO\Downloads\cleaned_autos.csv",encoding="Latin-1")
```

#1) Can you tell me No of Vehicles by Brand Available on ebay for sale with the help of visualization

```python
width=10
height=10
sns.set(rc={"figure.figsize":(width,height)})
a=sns.countplot(data2,y="Brand",order=data2["Brand"].value_counts().index)
for i in a.containers:
   a.bar_label(i)
plt.title("No of Vehicles by Brand Available for sale")
plt.show()
```

#2) What is the Average Price for vehicles based on the type of vehicle as well as 
#on the type of gearbox.Explain me with both numerical and visualization analysis

```python
d2=pd.pivot_table(data2,index=["VehicleType","Gearbox"],values=[ 'price'])
d2
```

```python
width=15
height=6
sns.set(rc={"figure.figsize":(width,height)})
a=sns.barplot(data=d2,x="VehicleType",y="price",hue="Gearbox")
for i in a.containers:
    a.bar_label(i)
plt.title("Average price of vehicles based on VehicleType and Gearbox Type")    
plt.show()
```
