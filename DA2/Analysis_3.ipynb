#2) What is the Average Price of vehicle by fuel type and gearbox type.Give a plot

```python
d2=pd.pivot_table(data2,index=["FuelType","Gearbox"],values=[ 'price'])
d2
```

```python
width=15
height=5
sns.set(rc={"figure.figsize":(width,height)})
a=sns.barplot(data=d2,x="FuelType",y="price",hue="Gearbox")
for i in a.containers:
    a.bar_label(i)
plt.title("Average price of vehicle based on FuelType and Gearbox")
plt.show()
```

#3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
d1=pd.pivot_table(data2,index=["VehicleType","Gearbox"],values=[ 'PowerPS'])
```

```python
d1
```

```python
a=sns.barplot(data=d1,x="VehicleType",y="PowerPS",hue="Gearbox")
for i in a.containers:
   a.bar_label(i)
plt.title("Average power of vehicles based on vehicletype and gearbox type")
plt.show()
```

#4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
d3=pd.pivot_table(data2,index=["Brand"],columns="Gearbox",values=[ 'price'],aggfunc=np.mean)
```

```python
d3
```

```python
sns.set(rc={"figure.figsize":(15,15)})
sns.heatmap(data=d3,annot=True,fmt="3g").set_title("Mean Price of the vehicles based on Brand and Vehicle Type")
plt.show()
```
